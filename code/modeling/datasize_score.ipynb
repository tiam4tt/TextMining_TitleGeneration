```python
rouge_scores = [
    [0.5195, 0.3146, 0.4470, 0.4472], # 3
    [0.5693, 0.3515, 0.4811, 0.4811], # 35
    [0.5886, 0.3746, 0.4977, 0.4977], # 4
    [0.6125, 0.4035, 0.5278, 0.5278], # 5
    [0.6310, 0.4263, 0.5431, 0.5431], # 55
    [0.6477, 0.4498, 0.5662, 0.5662], # 6
    [0.6438, 0.4416, 0.5522, 0.5522], # 65
    [0.6627, 0.4687, 0.5696, 0.5696], # 70
    [0.6071, 0.4124, 0.5442, 0.5460], #75
    [0.5633, 0.3948, 0.5107, 0.5107] # 80
    ]
avg = [sum(scores)/4 for scores in rouge_scores]
data = {
    "Cosine Similarity": [0.3, 0.35, 0.4, 0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8],
    "Average ROUGE Score": [avg],
    "BERT Scores": [0.9053, 0.9090, 0.9116, 0.9164, 0.9190, 0.9200, 0.9225, 0.9217, 0.9190, 0.9087]
}

import numpy as np
import matplotlib.pyplot as plt

# Flatten avg since it's a list of one list
avg_rouge = avg

# Min-max scaling
def min_max_scale(arr):
    arr = np.array(arr)
    return (arr - arr.min()) / (arr.max() - arr.min())

scaled_rouge = min_max_scale(avg_rouge)
scaled_bert = min_max_scale(data["BERT Scores"])

# Plotting with new color palette and no grid/y label
plt.figure(figsize=(8, 5))
plt.plot(data["Cosine Similarity"], scaled_rouge, marker='o', color='#1f77b4', label='Average ROUGE Score')
plt.plot(data["Cosine Similarity"], scaled_bert, marker='s', color='#ff7f0e', label='BERT Score F1')
plt.xlabel('Cosine Similarity')
plt.ylabel('Min-Max Scaled Score')
# plt.title('ROUGE and BERT Scores vs Cosine Similarity')
plt.legend()
# plt.yticks([])
plt.savefig('rouge_bert_vs_cosine.png', dpi=300, bbox_inches='tight')
plt.show()
```
